```python
#6
ecommerce=['Myntra','Snapdeal','Flipkart','Amazon','Ajio']
q1_profit=[35,45,105,70,45]
q2_profit=[25,24,100,70,40]
q3_profit=[30,42,115,70,48]
q4_profit=[38,40,120,72,50]
plt.scatter(ecommerce,q1_profit,color='green')
plt.scatter(ecommerce,q2_profit,color='blue')
plt.scatter(ecommerce,q3_profit,color='pink')
plt.scatter(ecommerce,q4_profit,color='red')
plt.xlabel("Organization name")
plt.ylabel("Profit")
plt.title("Scatter Plot")
plt.show()
```

```python
#8
q1_profit=[35,45,105,70,45]
q2_profit=[25,24,100,70,40]
q3_profit=[30,42,115,70,48]
q4_profit=[38,40,120,72,50]

plt.figure(1,figsize=(10,10))

plt.subplot(2,2,1)
plt.bar(ecommerce,q1_profit)
plt.title('Quater1 Profit')

plt.subplot(2,2,2)
plt.bar(ecommerce,q2_profit)
plt.title("Quater2 profit")

plt.subplot(2,2,3)
plt.bar(ecommerce,q3_profit)
plt.title('Quater3 Profit')

plt.subplot(2,2,4)
plt.bar(ecommerce,q4_profit)
plt.title('Quater4 Profit')
```

```python
#9
month_sales=[1,2,3,4,5,6,7,8,9]
electronics=[8,11,9,8,10,11,12,9,10]
clothing=[8,6,8,7,5,8,7,8,6]
food=[2,4,5,6,2,4,5,2,3]
books=[4,5,3,3,6,6,2,6,5]
plt.stackplot(month_sales,books,food,clothing,electronics,colors=['pink','blue','red','cyan','green'])
plt.xlabel('Months')
plt.ylabel("Sales")
plt.title("Area chart for sales of different categories")
plt.legend(['books','food','clothing','electronics'])
```
