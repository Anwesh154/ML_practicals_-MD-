```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np

```

```python
df=sns.load_dataset('titanic')
```

```python
df.head(5)
```

```python
df.shape
```

```python
df.keys()
```

```python
df.describe()
```

```python
df.columns
```

```python
sns.boxplot(x=df['class'],y=df['fare'],data=df).set_title('Boxplot for fare with respect to class')
```

```python
#from the diagram more we are able to tell a story the more it is interesting
sns.violinplot(x='class',y='age',hue='sex',data=df).set_title('Violin plot for showing age with class and sex')
```

```python
sns.lineplot(y=df['fare'],x=df['embarked'],data=df).set_title('Line plot for fare and embarked')
```

```python
sns.countplot(y=df['alive'],data=df).set_title("Countplot for alive")
```

```python
sns.barplot(y=df['fare'],x=df['embarked'],data=df)
```

```python
sns.stripplot(x='class',y='fare',jitter=True,data=df).set_title("Strip plot")
```

```python
sns.swarmplot(x='class',y='fare',data=df).set_title("Swarmplot")
```

```python
sns.FacetGrid(df,col="class")
grid1=sns.FacetGrid(df,col='class')
grid1.map(plt.hist,"age")
```

```python
sns.scatterplot(y='sepal_length',x='petal_length',data=iris).set_title("Scatter plot for sepal_length and petal_length")
```

```python
sns.regplot(y='petal_width',x='petal_length',data=iris)
```

```python
sns.lmplot(y='sepal_width',x='petal_length',data=iris,order=2)
```

```python
sns.heatmap(iris[['petal_length','petal_width']])
```

```python
sns.heatmap(iris[['petal_length','petal_width']],annot=True,cmap="YlGnBu")
```

```python
sns.heatmap(iris.corr(),annot=True,cmap='RdYlBu')
```

```python
sns.pairplot(iris,vars=['sepal_length','sepal_width',"petal_length"],kind='reg')
```

```python
sns.pairplot(iris,vars=['sepal_length','sepal_width','petal_length'],hue='species',diag_kind='hist',palette='colorblind')
```
